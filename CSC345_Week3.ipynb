```
# CSC345 -- Week#3 -- mini-lab  -- Getting to know your data
#                     with built-in IRIS flower dataset          
# Ref:  https://www.w3resource.com/machine-learning/scikit-learn/iris/index.php
#
# Instruction   -- 1. Uncomment each section of the code.  Run and observe
#                     the result of each section of codes, one at a time 
#                 
#
#------ Step#0 --- Load iris dataset 
import seaborn as sns
iris = sns.load_dataset('iris')

#------ Step#1 ---- print some basic properties of Iris dataset
print("Shape of Iris data set: ")
print(iris.shape)
print("\nKeys of Iris dataset:")
print(iris.keys())
print ("\n Data Types:")
print(type(iris))
print("\nFirst 4 rows:")
print(iris.head(4))
```

```
#------- Step#2 ----- print some row info., and NaN values (missing data) 
print("\nChecking for missing values")
print(iris.info())
```

```
#------- Step#3 ------- print basic statistics of data set 
#                       (Means, Std, Percentiles, Min, Max, Count ...)
print("\n\nBasic statistics of dataset")
print(iris.describe())
```

```
#-------- Step#4 -------- count the number of rows in each specy/class
print("\n\nObservations of each species:")
print(iris['species'].value_counts()) 
```

```
#-------- Step#5 -------- Plot a general statistics of dataset
import matplotlib.pyplot as plt
iris.describe().plot(kind = "area",fontsize=16, figsize = (15,8), table = True, colormap="Accent")
plt.xlabel('Statistics',)
plt.ylabel('Value')
plt.title("General Statistics of Iris Dataset")
plt.show()
```

```
#---------- Step#6 ------ Plot a barplot of 3 species/classes
ax=plt.subplots(1,1,figsize=(10,8))
sns.countplot('Species',data=iris)
plt.title("Iris Species Count++")
plt.show()
```
